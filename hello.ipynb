## Etape 1 : Tirage aléatoire

```java
import java.util.Random;
// ....
Random hasard = new Random();
int alea = hasard.nextInt(6+1-1) + 1; // nombre entre 1 et 6
System.out.println("Résultat du dé : " + alea);
```

## Etape 2 - Affichage

```java
char u = (alea % 2 == 1) ? '*' : ' ';
char d = (alea >= 2) ? '*' : ' ';
char q = (alea >= 4) ? '*' : ' ';
char s = (alea >= 2) ? '*' : ' ';

System.out.println(q + "   " + d);
System.out.println(s + " " + u + " " + s);
System.out.println(d + "   " + q);
```
