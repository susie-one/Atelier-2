## Etape 1 : Tirage aléatoire

```java
import java.util.Random;

Random hasard = new Random();

int alea1 = hasard.nextInt(6+1-1) + 1; // nombre entre 1 et 6
int alea2 = hasard.nextInt(6+1-1) + 1; // nombre entre 1 et 6

System.out.println("Premier dé : " + alea1);
System.out.println("Deuxième dé : " + alea2);
```

## Etape 2 - Affichage

```java
// Calcul du premier dé
char u1 = (alea1 % 2 == 1) ? '*' : ' ';
char d1 = alea1 >= 2 ? '*' : ' ';
char q1 = alea1 >= 4 ? '*' : ' ';
char s1 = alea1 >= 6 ? '*' : ' ';

// Calcul du deuxième dé
char u2 = (alea2 % 2 == 1) ? '*' : ' ';
char d2 = alea2 >= 2 ? '*' : ' ';
char q2 = alea2 >= 4 ? '*' : ' ';
char s2 = alea2 >= 6 ? '*' : ' ';

// Affichage côte à côte avec un trait
System.out.printf("%c   %c | %c   %c\n", q1, d1, q2, d2);
System.out.printf("%c %c %c | %c %c %c\n", s1, u1, s1, s2, u2, s2);
System.out.printf("%c   %c | %c   %c\n", d1, q1, d2, q2);

// Somme des deux dés
int somme = alea1 + alea2;
System.out.println("Somme des deux dés = " + somme);
```
